## 1.2 Introduction<a id='2.2_Introduction'></a>

This step focuses on collecting data, organizing it, and making sure it's well defined. Specifically, we have two sets of data - the historical data and the recent data are spread into two sets. 

### 1.2.1 Recap Of Data Science Problem<a id='1.2.1_Recap_Of_Data_Science_Problem'></a>

The purpose of this data science project is to come up with a forecasting model for organizations such as the federal reserve or any private business that is interested in forecasting the inflation rate in the near future. The Federal Reserveâ€™s dual mandate requires monetary policy to aim for both maximum employment and price stability. [As the RMSEs show](https://www.clevelandfed.org/en/newsroom-and-events/publications/economic-trends/2011-economic-trends/et-20110107-the-accuracy-of-cpi-inflation-forecasts.aspx), the accuracy of inflation forecasts has varied a lot over time. Inflation forecasts were not very accurate in the 1970s and 1980s. Accuracy improved in the 1990s and began to fall again in the 2000s. 

This model is aimed to utilize disaggregate information to improve the forecast accuracy for the aggregated inflation rate. The disaggregated information has been proposed by [many researchers](https://www.ecb.europa.eu/pub/pdf/scpwps/ecbwp589.pdf) with a dynamic factor model to reduce the dimensionality. However, the findings are mixed. This project aims to identify a statistical model that can improve the dynamic factor model in accuracy.  

The final product of the project is to generate a inflation forecasting for the next 12 months in a highly automated manner.

## 1.5 Load The Historical CPI Data<a id='1.5_Load_The_Historical_CPI_Data'></a>

There are 15 disaggregated features to be imported from the raw data. First of all, we import all the data frames and clean them one by one. To achieve this, I create a function that does data importation and cleaning automatically. 

```python
# The supplied CSV data file is the "../data/raw_data" directory
# Get the names of each disaggregated CPI data set
names = []
for root, dirs, files in os.walk("../data/raw_data"):
    for filename in files:
        names.append(filename)

# Define a function that takes the name of the sector and return a dataframe that 
# contains the sector-specific CPI information     
def function(name):
    df = pd.read_excel('../data/raw_data/' + name)[10:]
    df.columns = list(df.iloc[0])
    df.drop(10, inplace = True)
    df.drop(['HALF1', 'HALF2'], axis = 1, inplace = True)
    df.set_index('Year', inplace = True)
    raw_cpi = [list(df.iloc[i]) for i in range(0,df.shape[0])]
    cpi = np.reshape(raw_cpi, (12 * df.shape[0]))
    date = [datetime(year=y, month=m, day=calendar.monthrange(y,m)[1]) for y in range(df.index[0],df.index[-1] + 1) for m in range(1,13)]
    return pd.DataFrame({'date':date, name[0:-5]:cpi}) 

# Merge all the sector CPI information into one data frame
dfs = [function(name) for name in names]
df_final = reduce(lambda left,right: pd.merge(left,right,on='date',how = 'left'), dfs)
df_final.dropna(inplace = True)
```

## 1.8 Compute Inflation Data<a id='1.8_Compute_Inflation_Data'></a>

The goal of the project is to forecast the annual inflation rate in the next twelve months. I compute the inflation rate over the periods of one month, two months,..., all the way to 12 months. To illustrate the point, let's say we are standing at the end of October 2020, and we are trying to forecast the annual inflation rate of December 2020. First, we will forecast the 2-month-ahead inflation rate, denoted as $\pi$, which is the average inflation rate in November and December. The forecast of annual inflation rate for December can be computed as $((1 + \pi)CPI_{Oct,2020} - CPI_{Dec, 2019})/CPI_{Dec, 2019}$. To save space, I only demonstrate the computation of one-month-ahead inflation and twelve-month-ahead inflation.

```python
def plot_inflation(ahead, df_final):
    df_inflation_ahead = df_final.set_index("date").pct_change(periods=ahead).dropna().reset_index()
    formatted_df = pd.melt(df_inflation_ahead,
                           ["date"],
                           var_name="sector",
                           value_name="inflation")
    df_inflation_ahead_tidy = formatted_df.sort_values(by=["date"])
    df_inflation_ahead_tidy.reset_index(inplace = True)
    df_inflation_ahead_tidy.drop('index', axis = 1, inplace = True)
    df_inflation_ahead_tidy.head()

    g = sns.FacetGrid(df_inflation_ahead_tidy, col="sector", col_wrap=3)
    g.map(sns.lineplot, "date", "inflation")
    g.fig.set_size_inches(12,4)
```

## 1.8.2 Stationarity Twelve-Month-Ahead Inflation<a id='1.8.2_Stationarity_Twelve-Month-Ahead_Inflation'></a>

The graphs above show that the twelve-month-ahead inflation rate is stationary. We can run an augmented dickey-fuller test to show that the twelve-month-ahead inflation rate is indeed stationary.

```python
from statsmodels.tsa.stattools import adfuller
df_inflation_ahead_12 = df_final.set_index("date").pct_change(periods=12).dropna().reset_index()
sectors = list(df_inflation_ahead_12.columns)[1:-1]
p_vals = [round(adfuller(df_inflation_ahead_12[col])[1], 3) for col in sectors]
pd.DataFrame({'sector':sectors, 'p_val':p_vals})
```

The table above shows that the twelve-month-ahead inflation rates are mostly stationary processes (except for medical care services, commodities less food and energy commodities, and medical care commodities). 

## 1.8.3 One-Month-Ahead Inflation<a id='1.8.3_One-Month-Ahead_Inflation'></a>


```python
plot_inflation(1, df_final[['date','All items', 'Electricity', 'Commodities less food and energy commodities', 
                             'Shelter', 'Transportation services', 'Medical care services']])
```

The graphs above show that the one-month-ahead inflation rate is cyclical. We can run an augmented dickey-fuller test to test that whether the one-month-ahead inflation rate is stationary.

```python
from statsmodels.tsa.stattools import adfuller
df_inflation_ahead_1 = df_final.set_index("date").pct_change(periods=1).dropna().reset_index()
sectors = list(df_inflation_ahead_1.columns)[1:-1]
p_vals = [round(adfuller(df_inflation_ahead_1[col])[1], 3) for col in sectors]
pd.DataFrame({'sector':sectors, 'p_val':p_vals})
```

A time series with cyclic behaviour (but with no trend or seasonality) can be stationary. This is because the cycles are not of a fixed length, so before we observe the series we cannot be sure where the peaks and troughs of the cycles will be.

```python
# define useful function to create scatterplots of ticket prices against desired columns
def scatterplots(columns, df_inflation_ahead, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'All items', data=df_inflation_ahead, alpha=0.5)
        ax.set(xlabel=col, ylabel='Inflation')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

When the model trains, it only uses the previous ten years of data or 120 data points. This choice is made to avoid the biases introduced by the structural change across time.

## 3.2 Define Preprocessing Functions<a id='3.2_Define_Preprocessing_Functions'></a>

```python
# Preprocessing the model
def preprocess(ahead, df_final):
    df_inflation_ahead = df_final.set_index("date").pct_change(periods=ahead).dropna().reset_index()
    df_temp = df_inflation_ahead.set_index('date')
    df = df_temp.loc[df_temp.index > pd.to_datetime('1959-12-01')]
    # Create the X and y matrix
    dfperc = df * 100
    X = dfperc.drop('All items', axis = 1).values[:-ahead, :]
    y = dfperc.shift(-ahead)['All items'].values.reshape([-1, 1])[:-ahead, :]
    y = y.reshape(-1)
    # Illustrate time series cross validation 
    cv = TimeSeriesSplit(n_splits = 564, max_train_size = 120)
    return cv, X, y, dfperc

# Define the function that returns RMSE
def compute_RMSE(model, cv, X, y, ahead, dfperc, heldout = False):
	y_predictions = []
	y_truth = []
	time_index = []
	coefs = []
	for tr, tt in cv.split(X, y):
		if not heldout:
			if dfperc.index[tt + ahead] < pd.to_datetime('2005-01-01'):
				model.fit(X[tr], y[tr])
				#coefs.append(model.coef_)
				y_predictions.append(model.predict(X[tt]))
				y_truth.append(y[tt])
				time_index.append(dfperc.index[tt + ahead])
		else:
			if dfperc.index[tt + ahead] >= pd.to_datetime('2005-01-01'):
				model.fit(X[tr], y[tr])
				#coefs.append(model.coef_)
				y_predictions.append(model.predict(X[tt]))
				y_truth.append(y[tt])
				time_index.append(dfperc.index[tt + ahead])
	y_truth = np.array(y_truth).reshape(-1,1)
	y_predictions = np.array(y_predictions).reshape(-1,1)
	time_index = np.array(time_index).reshape(-1,1)
	RMSE = mean_squared_error(y_truth, y_predictions, squared = True)
	return RMSE, y_truth, y_predictions, time_index

# tune c and gamma so RMSE is small on the training set
def tune_c_gamma(ahead, df_final, c, gamma):
    cv, X, y, dfperc = preprocess(ahead, df_final)
    svr_c_vec = np.linspace(0.5*c, 1.5*c, 10)
    svr_RMSE_vec = np.array([compute_RMSE(SVR('rbf', C = c), cv, X, y, ahead, dfperc)[0] for c in svr_c_vec])
    svr_gamma_vec = np.linspace(0.5*gamma, 1.5*gamma, 10)
    svr_gamma_RMSE_vec = np.array([compute_RMSE(SVR('rbf', C = c, gamma = gamma), cv, X, y, ahead, dfperc)[0] for gamma in svr_gamma_vec])
    
    fig, ax = plt.subplots(1, 2)
    ax[0].plot(svr_gamma_vec, svr_gamma_RMSE_vec, marker = "o", linestyle = "--", linewidth = 0.5, markersize = 0.5, color = 'black')
    ax[0].set_ylabel("RMSE")
    ax[0].set_xlabel("gamma")

    ax[1].plot(svr_c_vec, svr_RMSE_vec, marker = "o", linestyle = "--", linewidth = 0.5, markersize = 0.5, color = 'black')
    ax[1].set_ylabel("RMSE")
    ax[1].set_xlabel("C")
    fig.set_size_inches([15, 4])
        
```

## 4 Deployment<a id='4_Deployment'></a>

## 4.1 Point Prediction<a id='4.1_Point_Prediction'></a>

```python
def point_prediction(year, month, C_vec, Gamma_vec, df_final, ahead, maximum_training_window):
    date = datetime(year=year, month=month, day=calendar.monthrange(year,month)[1]) 
    df_inflation_ahead = df_final.set_index("date").pct_change(periods=ahead).dropna().reset_index()
    df_temp = df_inflation_ahead.set_index('date')
    dfperc = df_temp * 100
    index = np.where(dfperc.index==date)[0][0]
    X_train = dfperc.drop('All items', axis = 1).iloc[index - maximum_training_window-ahead+1:index-ahead+1, :].values
    y_train = dfperc.iloc[index - maximum_training_window+1: index+1]['All items'].values.reshape(-1)
    X_test = dfperc.drop('All items', axis = 1).iloc[index, :].values
    model = SVR('rbf', C = C_vec[ahead-1], gamma = Gamma_vec[ahead-1])
    model.fit(X_train,y_train)
    return float(model.predict(X_test.reshape(-1, 14)))
```

## 4.2 Bootstrap Confidence Interval<a id='4.2_Bootstrap_Confidence_Interval'></a>

```python
def compute_CI(year, month, C_vec, Gamma_vec, df_final):
    CI_df = pd.DataFrame(columns = ['CI_0.05', 'CI_0.5', 'CI_0.95'])
    for ahead in range(1, 13):
        inflation_vec = [point_prediction(year, month, C_vec, Gamma_vec, df_final, ahead, maximum_training_window) for maximum_training_window in range(12,240)]
        CI_df.loc[ahead-1] = [np.quantile(inflation_vec, 0.05)] + [sum(inflation_vec)/len(inflation_vec)] + [np.quantile(inflation_vec, 0.95)]
    return CI_df
```

## 4.3 Convert to Annual Inflation Rate<a id='4.3_Convert_to_Annual_Inflation_Rate'></a>

```python
def convert_to_annual(forecast, year, month, df_final):
    date = datetime(year=year, month=month, day=calendar.monthrange(year,month)[1]) 
    i = np.where(df_final.date==date)[0][0]
    CPI_mid = df_final.iloc[i, :]['All items']
    CPI_start = [df_final.iloc[i-(11-ahead), :]['All items'] for ahead in range(0,12)]
    CPI_end = CPI_mid * (1 + forecast/100)
    forecast_annual = pd.DataFrame(columns = ['CI_0.05', 'CI_0.5', 'CI_0.95'])
    for i in range(0,12):
        forecast_annual.loc[i] = (CPI_end.loc[i] - CPI_start[i])/CPI_start[i] * 100
    return forecast_annual
```

## 4.4 Save Data<a id='4.4_Save_Data'></a>

```python
def compute_inflation_ci(year, month, C_vec, Gamma_vec, df_final):
    forecast = compute_CI(year, month, C_vec, Gamma_vec, df_final)
    forecast_annual = convert_to_annual(forecast, year, month, df_final)
    df_inflation_ahead = df_final.set_index("date").pct_change(periods=12).dropna().reset_index()
    df_temp = df_inflation_ahead.set_index('date')  
```

```python
higher = pd.read_csv('../data/generated_data/higher.csv')
middle = pd.read_csv('../data/generated_data/middle.csv')
lower = pd.read_csv('../data/generated_data/lower.csv')
actual = pd.read_csv('../data/generated_data/actual_inflation.csv')


higher['date'] = pd.to_datetime(higher['date'])
middle['date'] = pd.to_datetime(middle['date'])
lower['date'] = pd.to_datetime(lower['date'])
actual['date'] = pd.to_datetime(actual['date'])
```

```python
from dateutil.relativedelta import relativedelta
from matplotlib.pyplot import figure

year = 2020
month = 8

start_actual_date = datetime(year=year - 1, month=month, day=calendar.monthrange(year-1,month)[1]) 
start_forecast_date = datetime(year=year, month=month, day=calendar.monthrange(year,month)[1]) 
end_date = datetime(year=year + 1, month=month, day=calendar.monthrange(year+1,month)[1]) 


actual_ts = actual.loc[(actual.date >= start_actual_date) & (actual.date <= end_date)].set_index('date').rename(columns = {"All items":"Actual"}) 


middle_forc = pd.DataFrame({'date':[start_forecast_date + relativedelta(months=+m) for m in range(1, 13)], 'Forecast':middle.set_index('date').loc[start_forecast_date]}).set_index('date')
middle_forc.loc[start_forecast_date] = actual_ts.loc[start_forecast_date].values
middle_forc = middle_forc.iloc[np.arange(-1, len(middle_forc)-1)]


higher_forc = pd.DataFrame({'date':[start_forecast_date + relativedelta(months=+m) for m in range(1, 13)], 'Forecast':higher.set_index('date').loc[start_forecast_date]}).set_index('date')
higher_forc.loc[start_forecast_date] = actual_ts.loc[start_forecast_date].values
higher_forc = higher_forc.iloc[np.arange(-1, len(higher_forc)-1)]


lower_forc = pd.DataFrame({'date':[start_forecast_date + relativedelta(months=+m) for m in range(1, 13)], 'Forecast':lower.set_index('date').loc[start_forecast_date]}).set_index('date')
lower_forc.loc[start_forecast_date] = actual_ts.loc[start_forecast_date].values
lower_forc = lower_forc.iloc[np.arange(-1, len(lower_forc)-1)]



ax = plt.figure(figsize=(14,4))
plt.plot(middle_forc.index, middle_forc.Forecast, c = 'black', ls = '--') # plotting t, a separately 
plt.plot(higher_forc.index, higher_forc.Forecast, c = 'grey')
plt.plot(lower_forc.index, lower_forc.Forecast, c = 'grey')
plt.fill_between(higher_forc.index, higher_forc.Forecast, lower_forc.Forecast, color = "grey", label = "Prediction")
plt.plot(actual_ts.index, actual_ts.Actual, color='mediumvioletred', linewidth = 2, label = "Actual")
plt.axvline(start_forecast_date, ls = '--')
plt.ylim(-4, 6)

for i in range(-4, 6):
    plt.axhline(i, color = 'grey', lw = 1, ls = '--', alpha = 0.5)

plt.ylabel('Inflation (%)', fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.title(f"Predict Inflation From {year}-{month}-{calendar.monthrange(year,month)[1]}", fontsize=20)
ax.legend(loc = "center right", fontsize=12)
plt.show()
```
